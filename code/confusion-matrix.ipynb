```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Re-assign the confusion matrix values for demonstration purposes
tn, fp, fn, tp = 147, 52, 60, 141

# Creating the confusion matrix again with numpy
cm = np.array([[tn, fp], [fn, tp]])

# Plotting the confusion matrix using seaborn
sns.set(style="white", font_scale=1.4)   # Set style to have a white background for the plot
plt.figure(figsize=(8, 6))  # Set the figure size for better readability
heatmap = sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False)

# Adding labels for clarity
plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')
plt.xticks([0.5, 1.5], ['Conservative', 'Liberal'])
plt.yticks([0.5, 1.5], ['Conservative', 'Liberal'])

plt.savefig('confusion_matrix.png', bbox_inches='tight', dpi=300)

plt.show()

```
